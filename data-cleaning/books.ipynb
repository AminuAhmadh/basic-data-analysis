```python
# Percentage and count of missing values per row
for col in df.columns:
    nulls = df[col].isna().sum()
    pct = (nulls / df.shape[0]) * 100
    print('{} - {}%'.format(col, round(pct)))
    print(nulls)
    print()
```

```python
# Drop all the columns that has 90% or more missing data
df.drop(['Edition Statement', 'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver'], axis=1, inplace=True)
```
