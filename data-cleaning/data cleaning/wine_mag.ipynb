```python
# remove the "Unnamed 0" column - unnecessary column
df = pd.read_csv('Datasets\\winemag-data-130k-v2.csv', index_col=0)
df.head()
```

```python
# Select all the Qualitative columns and Numeric columns
qualitative = df.select_dtypes(exclude=[np.number])
qualitative
```

```python
numeric_dtype = df.select_dtypes(include=[np.number])
numeric_dtype
```

```python
# Get the know the number of rows and columns
df.shape
```

```python
# check out missing data
df.isnull().sum()
```

```python
# Drop the rows where countries data is missing
df['country'].dropna(inplace=True, axis=0)
```

```python
# Drop the missing data in variety col
df['variety'].dropna(inplace=True, axis=0)
```

```python
# reveal the rows with null values in designation column
df.loc[df['designation'].isnull()].head()
```

```python
# fill the missing values in designation with 'No Designation' - Missing values
df.designation.fillna('No Designation', inplace=True)
```

```python
# fill the missing values in price with the average - Missing values
df.price.fillna(df.price.mean, inplace=True)
```

```python
# Drop the rows with missing values in province columns - Missing values
df['province'].dropna(inplace=True, axis=0)
```

```python
# fill the NaN values in region 1 with 'unknown'
df['region_1'].fillna('unknown', inplace=True)
```

```python
# find the percent of the missing values in region 2 column
missing_values_count = df['region_2'].isnull().sum()
total_count = df.shape[0]
print(np.round(missing_values_count/total_count))
```

```python
# Drop the entire column region_2 - Missing data
df.drop('region_2', inplace=True, axis=1)
```

```python
# Let's remove the entire column
df.drop('taster_name', inplace=True, axis=1)
    
```

```python
# if taster twitter handle data is missing then there're probably not using twitter.
df['taster_twitter_handle'].fillna('No Twitter', inplace=True)
```
