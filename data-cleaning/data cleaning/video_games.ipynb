```python
import pandas as pd 
import numpy as np 
```

```python
vg = pd.read_csv('Datasets\\vgsales-12-4-2019-short.csv')
vg.head()
```

```python
# Lets start with setting the rank column as the index
vg.set_index('Rank', inplace=True)
```

```python
# 55792 rows and 15 columns
vg.info()
```

```python
# count how many NA's are in the colums
vg.isna().sum()
```

```python
# drop the any column that has na values above 40K
to_drop = ['Critic_Score', 'User_Score', 'Total_Shipped', 'NA_Sales', 'PAL_Sales', 'JP_Sales']
vg.drop(to_drop, axis=1, inplace=True)
```

```python
vg.head()
```

```python
# data types of the column
vg.dtypes
```

```python
# front fill all the na values
vg.fillna(method='ffill', inplace=True)
```

```python
vg.tail()
```

```python
# drop duplicates
vg.drop_duplicates(inplace=True)
```

```python
# No duplicates
vg.loc[vg.duplicated()]
```

```python
vg.to_csv('cleaned-vg-data')
```
