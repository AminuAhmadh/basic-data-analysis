```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
df = pd.read_csv('datasets//netflix_titles.csv')
```

```python
df.head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], infer_datetime_format=True)
```

```python
plt.title('pie chart for the type of shows')
plt.pie(df.type.value_counts(), labels=df.type.value_counts().index, autopct='%1.0f%%');
```

```python
df.columns
```

```python
cat_variable = df[['director', 'cast', 'country', 'rating','listed_in']]
```

```python
for col in cat_variable.columns:
    cols = cat_variable[col].value_counts()[:11]
    print(f'{col} Bar Chart')
    sns.barplot(x=cols.values, y=cols.index)
    plt.show()
```

```python
df.columns
```

```python
num_cols = df.select_dtypes(include=[np.number])
for col in num_cols.columns:
    print(f'Boxplot for {col} column')
    sns.boxplot(x=num_cols[col])
    plt.show()
```

```python
for col in num_cols.columns:
    print(f'Histogram for {col} column')
    sns.displot(x=num_cols[col])
    plt.show()
```

```python
for col in num_cols.columns:
    print(f'Kernal Density Plot for {col} column')
    sns.kdeplot(x=num_cols[col], shade=True)
    plt.show()
```

```python
sns.scatterplot(x='num_of_cast', y='num_of_directors', hue='type', data=df)
plt.show()
```
