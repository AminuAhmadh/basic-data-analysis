```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the data file
df = pd.read_csv('Data Files/cleaned-vg-data')
```

```python
df
```

```python
# the data isn't cleaned. so let's fix that before proceeding
df.isna().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
plt.figure(figsize=(12,6))
plt.title('Most common Genre')
sns.barplot(y=df['Genre'].value_counts().index, x=df['Genre'].value_counts().values)
```

```python
# Game release year
plt.figure(figsize=(16, 9))
plt.ylabel('Number of games released')
plt.xlabel('Year of game release')
years = df['Year'].value_counts()
sns.lineplot(y=years.values, x=years.index)
```

```python
df.columns
```

```python
plt.title('distribution of release year of games')
plt.hist(df.Year);
```

```python
plt.title('Distribution of global sales')
plt.hist(df.Global_Sales);
```

```python
plt.title('Distribution of other sales')
plt.hist(df.Other_Sales);
```

```python
platforms = df.Platform.value_counts()
plt.figure(figsize=(12,6))
plt.title('Which platform has the highest users/number of games')
sns.barplot(y=platforms.index[:11], x=platforms.values[:11])
```

```python
df.columns
```

```python
pltf = df.loc[(df.Platform == 'PC')  | (df.Platform == 'PS2')]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Global_Sales', y='Other_Sales', hue='Platform', data=pltf)
```

```python
sns.boxplot(x=df.Year);
```

```python
plt.figure(figsize=(12,6))
plt.title('Developers')
devs = df['Developer'].value_counts()[:20]
sns.barplot(x=devs.values, y=devs.index)
```

```python
plt.figure(figsize=(12,6))
plt.title('Publishers')
devs = df['Publisher'].value_counts()[:20]
sns.barplot(x=devs.values, y=devs.index)
```
