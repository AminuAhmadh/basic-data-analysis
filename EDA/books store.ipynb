```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('Data Files//cleaned book store data')
```

```python
df.head()
```

```python
pop = df['Place of Publication'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.title('Place Of Publication')
sns.barplot(x=pop.values[:12], y=pop.index[:12])
```

```python
pub = df['Publisher'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.title('Publishers')
sns.barplot(x=pub.values[:12], y=pub.index[:12])
```

```python
author = df['Author'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.title('Authors')
sns.barplot(x=author.values[:12], y=author.index[:12])
```

```python
cb = df['Contributors'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.title('Contributors')
sns.barplot(x=cb.values[:12], y=cb.index[:12])
```
